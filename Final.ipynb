```python
import csv

# 存储结果的列表
result = []

# 读取CSV文件
with open('/Users/shenzitong/Desktop/AAAI/收集的数据/texts_normal.csv', 'r', encoding='utf-8') as file:
    reader = csv.reader(file)
    next(reader)  # 跳过标题行
    for row in reader:
        # 将每一行转换为字符串格式
        text, label = row
        result.append(f'"{text}",{label}')

# 打印结果
for item in result:
    print(item)
```

```python
import jieba

# 停用词列表
stopwords = set()
with open('/Users/shenzitong/Desktop/AAAI/stopwords-master/baidu_stopwords.txt', 'r', encoding='utf-8') as f:
    stopwords = set(line.strip() for line in f)

# 使用 jieba 进行分词并去除停用词和空行
s_l_cut = []
for text in result:
    words = jieba.cut(text)
    filtered_words = [word for word in words if word not in stopwords and word.strip()]
    s_l_cut.append(' '.join(filtered_words))

# 将列表中的所有元素合并为一个字符串
combined_text = [' '.join(s_l_cut)]
# 将合并后的结果保存到本地文件
with open('/Users/shenzitong/Desktop/normal_分词后.txt', 'w', encoding='utf-8') as file:
    file.write(combined_text[0])
```

```python
from sklearn.feature_extraction.text import CountVectorizer

#初始化CountVectorizer，token_pattern后可以自定义提取单词的规则，不写则默认取纯字母单词，这里取的是字母、数字和-的组合
count_vect = CountVectorizer(token_pattern = r'[\u4e00-\u9fff]+')

#这里传入的参数必须是iteratable的，比如这里的list
X_train_counts = count_vect.fit_transform(combined_text)

from sklearn.feature_extraction.text import TfidfTransformer

tf_idf_transformer = TfidfTransformer()
tf_idf_matrix = tf_idf_transformer.fit_transform(X_train_counts) #传入bag-of-words矩阵即可

# 获取特征名称
feature_names = count_vect.get_feature_names_out()

# 将tf_idf_matrix转换为DataFrame
tf_idf_df = pd.DataFrame(tf_idf_matrix.toarray(), columns=feature_names)

# 保存到CSV文件
tf_idf_df.to_csv('/Users/shenzitong/Desktop/tf_idf_matrix_normal.csv', index=False, encoding='utf-8-sig')
```

```python
import pandas as pd

def merge_csv_files(scam_csv_path, normal_csv_path, output_csv_path):
    # 读取两个CSV文件
    df_scam = pd.read_csv(scam_csv_path, encoding='utf-8-sig')
    df_normal = pd.read_csv(normal_csv_path, encoding='utf-8-sig')
    
    # 添加标签列
    df_scam['label'] = 'scam'
    df_normal['label'] = 'normal'
    
    # 合并两个DataFrame
    merged_df = pd.concat([df_scam, df_normal], ignore_index=True)
    
    # 保存到CSV文件
    merged_df.to_csv(output_csv_path, index=False, encoding='utf-8-sig')

# 使用函数
merge_csv_files(
    scam_csv_path='/Users/shenzitong/Desktop/tf_idf_matrix_scam.csv',
    normal_csv_path='/Users/shenzitong/Desktop/tf_idf_matrix_normal.csv',
    output_csv_path='/Users/shenzitong/Desktop/tf_idf_matrix_combined.csv')
```

```python
import pandas as pd
import numpy as np

# 读取CSV文件并转换为数值
df = pd.read_csv('/Users/shenzitong/Desktop/tf_idf_matrix_combined.csv', encoding='utf-8-sig')
df = df.apply(pd.to_numeric, errors='coerce')

# 假设第一行是scam的TF-IDF，第二行是normal的TF-IDF
words = df.columns
scam_tfidf = df.iloc[0].fillna(0).values
normal_tfidf = df.iloc[1].fillna(0).values

# 计算差异
tfidf_diff = scam_tfidf - normal_tfidf

# 获取差异最大的前50个词
top_n = 50
top_indices = np.argsort(tfidf_diff)[-top_n:]

# 提取并排序
top_words = [(words[i], tfidf_diff[i]) for i in top_indices]
top_words_sorted = sorted(top_words, key=lambda x: x[1], reverse=True)

# 将结果保存到CSV文件
output_df = pd.DataFrame(top_words_sorted, columns=['word', 'difference'])
output_df.to_csv('/Users/shenzitong/Desktop/top_50_words.csv', index=False, encoding='utf-8-sig')

# 打印结果
for word, diff in top_words_sorted:
    print(f"{word}: {diff}")
```

```python
#Combine所有txt到一个dataframe中
import pandas as pd
import os

# Define directories
scam_dir = '/Users/shenzitong/Desktop/AAAI/收集的数据/诈骗文字升级版'
normal_dir = '/Users/shenzitong/Desktop/AAAI/收集的数据/正常对话'

# Function to read text files into a DataFrame
def read_text_files(directory, label):
    texts = []
    for filename in os.listdir(directory):
        if filename.endswith('.txt'):
            filepath = os.path.join(directory, filename)
            with open(filepath, 'r', encoding='UTF-8') as file:
                text = file.read().strip()
                texts.append({'text': text, 'label': label})
    return pd.DataFrame(texts)

# Read scam and normal texts
scam_df = read_text_files(scam_dir, 1)  # Label as 1 for scam
normal_df = read_text_files(normal_dir, 0)

# Combine the DataFrames
combined_df = pd.concat([scam_df, normal_df], ignore_index=True)

# Write to CSV
combined_df.to_csv('/Users/shenzitong/Desktop/AAAI/收集的数据/texts_all.csv', index=False, header=['文本', '标签'], encoding='UTF-8-sig')
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split, cross_val_predict
from sklearn.svm import SVC
from sklearn.metrics import precision_recall_fscore_support, accuracy_score
from sklearn.model_selection import StratifiedKFold

# 读取CSV文件
df = pd.read_csv('/Users/shenzitong/Desktop/top_50_words.csv', encoding='utf-8-sig')
# 提取前50个词
top_words_sorted = df['word'].head(50).tolist()

# 初始化 CountVectorizer，指定词汇表
count_vect = CountVectorizer(vocabulary=top_words_sorted, token_pattern=r'[\u4e00-\u9fff]+')

# 向量化合并的文本
X_counts_combined = count_vect.fit_transform(s_l_cut)

# 读取 CSV 文件
df = pd.read_csv('/Users/shenzitong/Desktop/segmented_texts.csv', encoding='UTF-8-sig')

# 提取标签列
y = df['Label'].tolist()

# 划分训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X_counts_combined, y, test_size=0.2, random_state=42)

# 创建SVM模型
svm = SVC(kernel='linear')

# 交叉验证
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# 预测所有数据集
y_pred = cross_val_predict(svm, X_counts_combined, y, cv=cv)

# 计算指标
precision, recall, f1, support = precision_recall_fscore_support(y, y_pred)
accuracy = accuracy_score(y, y_pred)

# 输出每个标签的平均指标
unique_labels = sorted(set(y))
for i, label in enumerate(unique_labels):
    print(f"Label {label}:")
    print(f"  Precision: {precision[i]:.3f}")
    print(f"  Recall:    {recall[i]:.3f}")
    print(f"  F1 Score:  {f1[i]:.3f}")
    print(f"  Support:   {support[i]}")

# 输出整体平均准确率
print(f"Overall Accuracy: {accuracy:.3f}")
```

```python
from sklearn.neighbors import KNeighborsClassifier
# 训练KNN模型
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train, y_train)

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# 存储每轮指标
all_precisions, all_recalls, all_f1s, all_accuracies = [], [], [], []

for fold, (train_idx, test_idx) in enumerate(cv.split(X_counts_combined, y), 1):
    X_train, X_test = X_counts_combined[train_idx], X_counts_combined[test_idx]
    y_train, y_test = np.array(y)[train_idx], np.array(y)[test_idx]
    
    knn_model.fit(X_train, y_train)
    y_pred = knn_model.predict(X_test)
    
    precision, recall, f1, _ = precision_recall_fscore_support(y_test, y_pred, average=None)
    accuracy = accuracy_score(y_test, y_pred)
    
    all_precisions.append(precision)
    all_recalls.append(recall)
    all_f1s.append(f1)
    all_accuracies.append(accuracy)
    
    print(f"Fold {fold}:")
    for i, label in enumerate(sorted(set(y))):
        print(f"  Label {label}:")
        print(f"    Precision: {precision[i]:.3f}")
        print(f"    Recall:    {recall[i]:.3f}")
        print(f"    F1 Score:  {f1[i]:.3f}")
    print(f"  Accuracy: {accuracy:.3f}")
    print()

# 输出整体平均
mean_precision = np.mean(all_precisions, axis=0)
mean_recall = np.mean(all_recalls, axis=0)
mean_f1 = np.mean(all_f1s, axis=0)
mean_accuracy = np.mean(all_accuracies)

print("Overall Averages:")
for i, label in enumerate(sorted(set(y))):
    print(f"  Label {label}:")
    print(f"    Precision: {mean_precision[i]:.3f}")
    print(f"    Recall:    {mean_recall[i]:.3f}")
    print(f"    F1 Score:  {mean_f1[i]:.3f}")
print(f"Overall Accuracy: {mean_accuracy:.3f}")
# # 交叉验证
# cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# # 预测所有数据集
# y_pred = cross_val_predict(svm, X_counts_combined, y, cv=cv)

# # 计算指标
# precision, recall, f1, support = precision_recall_fscore_support(y, y_pred, average=None)
# accuracy = accuracy_score(y, y_pred)

# # 输出每个标签的平均指标
# unique_labels = sorted(set(y))
# for i, label in enumerate(unique_labels):
#     print(f"Label {label}:")
#     print(f"  Precision: {precision[i]:.3f}")
#     print(f"  Recall:    {recall[i]:.3f}")
#     print(f"  F1 Score:  {f1[i]:.3f}")
#     print(f"  Support:   {support[i]}")

# # 输出整体平均准确率
# print(f"Overall Accuracy: {accuracy:.3f}")
```
